## Introduction to Tree Data Structure
A tree is a hierarchical data structure used to organize and represent data in a parent–child relationship.
It consists of nodes, where the topmost node is called the root, and every other node can have one or more child nodes.

### Binary Tree
Trees which has atmost 2 child node or 0 child node

### Types of Binary Tree
1. Full Binary Tree : Tree having 2 or 0 children
2. Complete Binary Tree : (i) All level must be full except last level (ii) The last level has all nodes to the extreme left
3. Perfect Binary Tree : (i) All leaf node are at same level (ii) All non-leaf nodes must have 2 childrens
4. Balanced Binary Tree :  Hieght difference between left and right subtree at any node must be of max 1
5. Degenerate Binary Tree : Every node has one child

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

# Create nodes
drink = Node("drink")
hot = Node("hot")
cold = Node("cold")
tea = Node("tea")
coffee = Node("coffee")
cola = Node("cola")
fanta = Node("fanta")

# Build the tree
drink.left = hot
drink.right = cold

hot.left = tea
hot.right = coffee

cold.left = cola
cold.right = fanta

# Access values
print(drink.val)                 # drink
print(drink.left.val)            # hot
print(drink.left.right.val)      # coffee
```
