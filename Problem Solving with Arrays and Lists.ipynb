## Reverse a Number and Check Whether It Is a Palindrome
Given an integer number, write a program to reverse the number and check whether it is a palindrome or not.

```python
#1.DSA PRoblem 1
number= 1441
original_num=number
rev = 0
while (number > 0):
    rem = number % 10
    rev = rev * 10 + rem
    number = number // 10
if (rev == original_num):
    print(f"the {rev} is palindorm.")
else:
    print ("it is not palindrom")


```
